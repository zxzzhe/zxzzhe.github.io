```python
def pai(N):
    a=[]
    a.append(2)
    for i in range(3,N):
        for j in range(len(a)):
            if i%a[j]==0:
                break
            if i%a[j]!=0 and j==len(a)-1:
                a.append(i)
    m=len(a)
    return m/N
print('N=10,100,1000,10000,100000时的值分别为:')
print(pai(10),pai(100),pai(1000),pai(10000),pai(100000),sep = '  ')
```
