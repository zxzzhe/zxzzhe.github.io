```python
import math
import random
import matplotlib.pyplot as plt
N=10000
t=[random.uniform(0,2*math.pi) for i in range(N)]
move=[random.randrange(-1,2,2) for i in range(N)]
move1=[[move[i]*math.cos(t[i]),move[i]*math.sin(t[i])] for i in range(N)]
x=[]
y=[]
consum=[0,0]
for item in move1:
    consum[0]+=item[0]
    consum[1]+=item[1]
    x.append(consum[0])
    y.append(consum[1])
plt.plot(x,y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
def fi(n):
    if n==0:
        fi_list=[0]
    elif n==1:
        fi_list=[0,1]
    else:
        f_0,f_1=0,1
        fi_list=[]
        fi_list.append(f_0)
        fi_list.append(f_1)
        for i in range(n-2):
            fi_num=f_0+f_1
            fi_list.append(fi_num)
            f_0,f_1=f_1,fi_num
    return fi_list

def center(f_list):
    x_n, y_n = 0, 0
    o_x_array, o_y_array = np.array([x_n,x_n]), np.array([y_n,y_n])
    for n in np.arange(2,len(f_list)):
        y_n=y_n+n%2*f_list[n-1]*(-1)**(1+((n+1)%2+n+1)/2)
        x_n=x_n+(n+1)%2*f_list[n-1]*(-1)**(1+((n+1)%2+n+3)/2)
        o_x_array = np.append(o_x_array, x_n)
        o_y_array = np.append(o_y_array, y_n)
    return o_x_array, o_y_array
num = 7
f_list=fi(num)
x0_array,y0_array=center(f_list)
r=fi(num+1)[1::]
angle1,angle2=0.5*np.pi,np.pi
for n in np.arange(len(f_list)):
    t=np.arange(angle1,angle2,0.001)
    x=r[n]*np.sin(t)+x0_array[n]
    y=r[n]*np.cos(t)+y0_array[n]
    angle1+=0.5*np.pi
    angle2+=0.5*np.pi
    plt.plot(x,y,color='black')
    x1=x0_array[n]
    y1=y0_array[n]
    x3=x1+r[n]*(-1)**(1+(n%2+n+2)/2)
    y3=y1+r[n]*(-1)**(1+((n+1)%2+n+3)/2)
    x2,y2=x3,y1
    x4,y4=x1,y3
    rectangle1=[x1,x2,x3,x4,x1]
    rectangle2=[y1,y2,y3,y4,y1]
    plt.plot(rectangle1,rectangle2,color='black')
```
