```python
import numpy as np
import matplotlib.pyplot as plt
xi=np.random.uniform(low=0,high=1,size=10000)
plt.hist(xi,bins=100,density=1)
ave1=xi.mean()
var1=xi.std()
print('平均值：{}'.format(ave1))
print('方差{}'.format(var1**2))
```

```python
x=xi.cumsum()
n1=np.arange(1,len(x)+1)
x2=x/n1
plt.hist(x2,bins=100,density=1)
ave2=x2.mean()
var2=x2.std()
print('平均值：{}'.format(ave2))
print('方差{}'.format(var2**2))
```

```python
yi=np.sin(2*np.pi*xi)
plt.hist(yi,bins=100,density=1)
ave3=yi.mean()
var3=yi.std()
print('平均值：{}'.format(ave3))
print('方差{}'.format(var3**2))
```

```python
y=yi.cumsum()
n2=np.arange(1,len(y)+1)
z=y/n2
plt.hist(z,bins=100,density=1)
ave4=z.mean()
var4=z.std()
print('平均值：{}'.format(ave4))
print('方差{}'.format(var4**2))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
A1=1
A2=2
A3=3
A4=4
P=0.5
omiga=0.5
def deff_equation1(y_list,t,P,omiga):
    xita,x=y_list
    return np.array([x,A1*np.cos(xita)*np.cos(omiga*t)-np.sin(xita)-P*x])
def deff_equation2(y_list,t,P,omiga):
    xita,x=y_list
    return np.array([x,A2*np.cos(xita)*np.cos(omiga*t)-np.sin(xita)-P*x])
def deff_equation3(y_list,t,P,omiga):
    xita,x=y_list
    return np.array([x,A3*np.cos(xita)*np.cos(omiga*t)-np.sin(xita)-P*x])
def deff_equation4(y_list,t,P,omiga):
    xita,x=y_list
    return np.array([x,A4*np.cos(xita)*np.cos(omiga*t)-np.sin(xita)-P*x])
paras = (P, omiga)
step=0.01
t=np.arange(0.0,100, step)
track1=odeint(deff_equation1,(0.25*np.pi,0),t,args=paras)
track2=odeint(deff_equation2,(0.25*np.pi,0),t,args=paras)
track3=odeint(deff_equation3,(0.25*np.pi,0),t,args=paras)
track4=odeint(deff_equation4,(0.25*np.pi,0),t,args=paras)
plt.plot(track1[:,0], track1[:,1])
plt.plot(track2[:,0], track2[:,1])
plt.plot(track3[:,0], track3[:,1])
plt.plot(track4[:,0], track4[:,1])
plt.title("周期驱动震动")
plt.xlabel("$θ$")
plt.ylabel("单摆$\omega$")
```
