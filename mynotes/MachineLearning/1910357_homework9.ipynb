```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('faithful.csv')
data.plot.scatter(x='eruptions',y='waiting')
```

```python
df=data.copy()
df['x*y']=df['eruptions']*df['waiting']
sum_xy=df['x*y'].sum()
df['x*x']=df['eruptions']*df['eruptions']
sum_xx=df['x*x'].sum()
ave_x=df['eruptions'].mean()
ave_y=df['waiting'].mean()
n=len(df)
k=(sum_xy-n*ave_x*ave_y)/(sum_xx-n*ave_x*ave_x)
b=ave_y-k*ave_x
print('waiting={}*eruptions+{}'.format(k,b))
```

```python
df1=df.sort_values(by='eruptions')
df1['class']=df['eruptions']<3.25
result=df1.groupby('class').mean()
result
print('两团数据各自的中心位置为:\n eruptions={}\n waiting={}\n和\n eruptions={}\n waiting={}'.format(result.iloc[0,0],result.iloc[0,1],result.iloc[1,0],result.iloc[1,1]))
```
